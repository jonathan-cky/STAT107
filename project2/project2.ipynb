# Section 4: Data Science

Now that we have seen the closing price graphs for both companies, we want to see if the two companies have created a strategic partnership and if this partnership has affected their stock price. However, since both companies are in the techonological field, understand that both companies will have a high correlation even without the partnership. For this section we want to find the correlation between the two companies' closing prices after 2014. You may start off by creating a new data frame for closing prices.

```python
aaple_close = aapl_2014_after["Close"].reset_index(drop=True)
tsm_close = tsm_2014_after["Close"].reset_index(drop=True)

aaple_close = pd.DataFrame({'Close_aaple': aaple_close})
tsm_close = pd.DataFrame({'Close_tsm': tsm_close})

df_cor = pd.concat([aaple_close, tsm_close], axis=1)
df_cor
```

##### After you get the two closing prices, find the correlation with the correlation function in python.

# Section 5: Overall Summary

Due to the high correlation between the two companies, whether due to the collaboration or similiar products. We are curious if we can predict the stock price of one company only by using the data(trading data we imported in section one) from the other. Run the code below and see if our hypothesis is correct. Again, we want to predict the closing price. To see if our predictions are accurate, we wil be finding the mean squared error of between our predictions and the actual price. Run the code below and see if our predictions are accurate.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

model = LinearRegression()
aapl_1997 = df_aaple[df_aaple["Date"] >= "1997-10-09"]
aapl_1997.reset_index(drop=True, inplace=True)
df_tsm["aaple_Close"] = aapl_1997["Close"]

test_df = df_tsm.sample(frac = 0.2)
train_df = df_tsm[~df_tsm.index.isin(test_df.index)]

independent_variables = ["Open", "High", "Low","Close", "Adj Close", "Volume"]
dependent_variable = "aaple_Close"

model.fit(train_df[independent_variables], train_df[dependent_variable])
predictions = model.predict( test_df[independent_variables] )

mse = mean_squared_error(test_df[dependent_variable], predictions)
print("Mean Squared Error:", mse)
```

##### Now lets see we can predict the closing stock price of aapl by only using the data from aapl. Then we will compare the mean squared error of using tsm's data to using aapl's data and see which one performs better.

```python

model = LinearRegression()

test_df = df_aaple.sample(frac = 0.2)
train_df = df_aaple[~df_aaple.index.isin(test_df.index)]

independent_variables = ["Open", "High", "Low", "Adj Close", "Volume"]
dependent_variable = "Close"

model.fit(train_df[independent_variables], train_df[dependent_variable])
predictions = model.predict( test_df[independent_variables] )

mse = mean_squared_error(test_df[dependent_variable], predictions)
print("Mean Squared Error:", mse)
```
